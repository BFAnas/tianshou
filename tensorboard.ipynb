```python
import pandas as pd
from tensorboard.backend.event_processing import event_accumulator

# Path to your TensorBoard log directory
log_dir = "/data/user/R901105/dev/log/HalfCheetah-v2/SafeRL/240129-121039"

# Initialize an Event Accumulator
ea = event_accumulator.EventAccumulator(log_dir,
    size_guidance={event_accumulator.SCALARS: 0})

# Load the data
ea.Reload()

# Extract scalar data
scalar_tags = ea.Tags()["scalars"]

# Prepare DataFrame
data = {}
for tag in scalar_tags:
    events = ea.Scalars(tag)
    steps = [event.step for event in events]
    values = [event.value for event in events]
    data[tag] = pd.Series(values, index=steps)

df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df = df[[col for col in df.columns if "Stats" in col]].dropna()
```

```python
df.to_csv("HC_NR2.csv")
```
